#Queries

```python
#Number of products in each category
query = '''
SELECT product_category_name, COUNT(product_id) AS product_number FROM products
GROUP BY product_category_name
'''
pd.read_sql(query, conn)

```

This information could be helpful when the company wants to know in which category they can focus their resources to optimize inventory

```python
#English translation JOIN
query = '''SELECT
products.*, product_category_name_translation.product_category_name_english
FROM products
JOIN product_category_name_translation ON product_category_name_translation.product_category_name=products.product_category_name
'''
pd.read_sql(query, conn)
```

Adding the english translation of products categories to the products table may be helpful for international employees and costumers


```python
#WHERE >1000
query = '''
SELECT * FROM order_payments
WHERE order_payments.payment_value > 1000
'''
pd.read_sql(query, conn)
```

Showing all payments where the value is greater than 1000 can help identify significant transactions.

```python
#Average freight value rounded by 2
query = '''
SELECT ROUND(AVG(order_items.freight_value),2) AS freight_value_avg
FROM order_items
'''
pd.read_sql(query, conn)
```


The company can use freight value to compare its shipping costs, find ways to save money, and work with carriers to secure better rates.

```python
#Order's status and delivery time
query = '''
SELECT order_items.shipping_limit_date, COUNT(orders.order_id) AS total_orders, orders.order_status, orders.order_estimated_delivery_date
FROM order_items
JOIN orders ON order_items.order_id=orders.order_id
GROUP BY orders.order_status
'''
pd.read_sql(query, conn)
```

Order status and delivery time are crucial for managing customer expectations, it allows much better communication with customers and helps identify areas for optimizing service levels.

```python
#Average number of installments across all orders
query = '''
SELECT AVG(payment_installments) AS avg_installments
FROM order_payments
'''
pd.read_sql(query, conn)
```

```python
#price of delivered orders grouped by installments number
query = '''
SELECT order_items.price, orders.order_status, order_payments.payment_installments
FROM order_items
JOIN orders ON order_items.order_id=orders.order_id
JOIN order_payments ON orders.order_id=order_payments.order_id
WHERE orders.order_status = 'delivered'
GROUP BY order_payments.payment_installments
'''
pd.read_sql(query, conn)
```

This data can be used to segment customers based on their payment preferences, also using this query as a template we can get information about other orders with another status.

```python
query = '''
SELECT COUNT(order_payments.order_id) AS unique_orders, COUNT(order_items.product_id) AS total_items, COUNT(order_payments.payment_sequential) AS total_payments
FROM order_payments
JOIN order_items ON order_payments.order_id=order_items.order_id
GROUP BY order_payments.order_id
HAVING total_payments > 5
'''
pd.read_sql(query, conn)
```

For orders with duplicate order_ids, each item in the order triggers a separate payment.
The total number of payments is equal to the number of items in the order.

```python
#Rating for the product category and total sold items
query = '''
SELECT
order_reviews.review_score, products.product_category_name, COUNT(order_items.product_id) AS total_sold
FROM order_items
JOIN order_reviews ON order_items.order_id=order_reviews.order_id
JOIN products ON order_items.product_id=products.product_id
GROUP BY order_reviews.review_score
'''
pd.read_sql(query, conn)
```

This query shows the most appreciated product categories by customers and how much products of each category was sold.

```python
#top 5 sold products
query='''
SELECT products.product_category_name, COUNT(order_items.product_id) AS total_sold
FROM order_items
JOIN products ON order_items.product_id=products.product_id
GROUP BY products.product_category_name
ORDER BY total_sold DESC
LIMIT 5
'''
pd.read_sql(query,conn)
```

The top 5 product categories with the most sold items can offer valuable insights for making smarter decisions about future purchases and marketing strategies. By recognizing which categories are performing best, the company can focus its efforts on keeping popular products well-stocked and visible, ensuring they get the attention they deserve.

```python
query = '''
SELECT
    COUNT(order_items.order_id) AS total_orders,
    GROUP_CONCAT(order_items.product_id) AS products,
    COUNT(order_items.product_id) AS total_products,
    SUM(order_items.price) AS total_price,
    order_payments.payment_type
FROM order_items
JOIN order_payments ON order_payments.order_id = order_items.order_id
GROUP BY order_items.order_id
HAVING total_price >= 100
'''
pd.read_sql(query,conn)
```

This query lists all the product IDs in that order (concatenated into a single string), and returns all orders, total price of orders and the payment type where total price is greater than 100.
This may be used to know the total price of each order even it has more than 1 sold product in the invoice and how that orders are paid.


```python
query = '''
SELECT
    COUNT(order_items.order_id) AS total_orders,
    GROUP_CONCAT(order_items.product_id) AS products,
    COUNT(order_items.product_id) AS total_products,
    SUM(order_items.price) AS total_price,
    order_payments.payment_type,
    orders.order_status
FROM order_items
JOIN order_payments ON order_payments.order_id = order_items.order_id
JOIN orders ON order_items.order_id=orders.order_id
GROUP BY order_items.order_id
HAVING orders.order_status = 'invoiced'
'''
#HAVING orders.order_status = 'approved'
#HAVING orders.order_status = 'shipped'
#HAVING orders.order_status = 'delivered'
#HAVING orders.order_status = 'unavailable'
#HAVING orders.order_status = 'canceled'
#HAVING orders.order_status = 'processing'
#or order_status may be not specified
pd.read_sql(query,conn)
```

We can use this query to know about payment process of orders with specific order status.

```python
import plotly
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

```python
#Analyzing the relationship between product price and freight value can reveal patterns in shipping costs relative to product price.

order_items_df = pd.read_sql_query("SELECT price, freight_value FROM order_items", conn)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='freight_value', data=order_items_df)
plt.title('Freight Value vs. Price')
plt.xlabel('Price')
plt.ylabel('Freight Value')
plt.show()
```

```python
#Calculating the total revenue per seller will show which sellers contribute the most to sales.

seller_revenue_df = pd.read_sql("SELECT order_items.price, order_items.seller_id FROM order_items", conn)
seller_revenue = seller_revenue_df.groupby('seller_id')['price'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
seller_revenue.plot(kind='bar', color='green')
plt.title('Top Sellers by Total Revenue')
plt.xlabel('Seller ID')
plt.ylabel('Total Revenue')
plt.show()
```

```python
query = "SELECT * FROM geolocation"
geo_df = pd.read_sql_query(query, conn)
geo_df.head()
```

```python
fig = px.sunburst(geo_df,
                  path=['geolocation_state', 'geolocation_city'],
                  title="Geolocations by State and City",
                  hover_data=['geolocation_zip_code_prefix'])

fig.show()
```

```python
#Top 10 Selling products
query = """SELECT  product_category_name_translation.product_category_name_english as Mostly_sold_product_type, SUM(price+freight_value) as Revenue FROM order_items
JOIN products ON order_items.product_id = products.product_id
JOIN product_category_name_translation ON products.product_category_name = product_category_name_translation.product_category_name
GROUP BY product_category_name_translation.product_category_name_english
ORDER BY
    Revenue DESC
Limit 10
;"""
pd.read_sql(query, conn)
df = pd.read_sql_query(query, conn)
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Mostly_sold_product_type', y='Revenue')

plt.title('Top 10 Selling Product Types by Revenue', fontsize=14)
plt.xlabel('Product Type', fontsize=12)
plt.ylabel('Revenue (in $)', fontsize=12)

plt.xticks(rotation=45)

plt.show()
```

```python
# Query for Seller City Order Analysis (Delivered Orders Only)
query_city = """
SELECT sellers.seller_city,
SUM(price) AS Sales,
COUNT(*) AS Count
FROM order_items
JOIN sellers ON sellers.seller_id = order_items.seller_id
JOIN orders ON orders.order_id = order_items.order_id
WHERE orders.order_status = 'delivered'
GROUP BY sellers.seller_city
ORDER BY SUM(price) DESC
LIMIT 10
"""
city_data = pd.read_sql(query_city, conn)

# Seaborn Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='seller_city', data=city_data, palette='mako')
plt.xlabel('Number of Delivered Orders', fontsize=12)
plt.ylabel('Seller City', fontsize=12)
plt.title('Top 10 Seller Cities by Number of Delivered Orders', fontsize=16)
plt.tight_layout()
plt.show()
```
